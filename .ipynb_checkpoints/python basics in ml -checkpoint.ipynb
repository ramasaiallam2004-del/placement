## Loops 

```python
#for Loop 

for i in range(5): 
   print(i) 
```

```python
 for i in range (2,11,2): 
     print(i) 
```

```python
marks = [70,80,90] 
for mark in marks: 
  print(mark) 
```

```python
#while Loop 
count = 1 
while count <= 5: 
  print(count) 
```
