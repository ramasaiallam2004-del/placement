## Creating array

```python
a = np.array([1,2,3,4]) # 1_d array
print(a)
```

```python
b = np.array([[1,2,3,4],[5,6,7,8]])  # 2-d array
print(b)
```

```python
a = np.array([0,0,0,0])
print(a)
```

```python
zero = (np.zeros((2,3)))
print(zero)
```

```python
print(np.ones((2,3)))
```

```python
print(np.arange(0,10,2))
```

```python
print(np.eye(3)) # identity  matrix
```

```python
rng = np.random.default_rng(seed=42)
print(rng.random((2,3)))
```

```python
print(rng.integers(1,10,size=(3,3)))
```

```python
print(rng.normal(loc=0, scale = 1, size =5))
```

## Array Atttibutes and data types

```python
M = np.array([[1,2,3,],[4,5,6,]])
M
```

```python
print(M.ndim)
```

```python
print(M.shape)
```

```python
print(M.size)
```

```python
print(M.dtype)
```

```python
ints = np.array([1,2,3], dtype=np.int32)
ints.dtype
```

```python
floats = np.array([1,2,3], dtype = np.float64)
floats.dtype
```

## indexing and slicing

```python
x = np.arange(10,20)
x
```

```python
print(x[4])
```

```python
print(x[:4])
```

```python
M = np.arange(1,13).reshape(3,4)
M
```

```python
print(M[1,2])
```

```python
print(M[1])
```

```python
print(M[:,0])
```

## Vector operation

```python
a = np.array([1,2,3,4])
b = np.array([10,20,30,40])
```

```python
print(a)
print(b)
```

```python
print(a+b)
```

```python
print(a*b)
```

```python
# boardcasting
```

```python
a = np.array([1,2,3])
b = 10
print(a+b)
```

```python
a = np.array([[1,2,3],
              [4,5,6],])
b = np.array([10,20,30])
```
